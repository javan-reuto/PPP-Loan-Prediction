```python
def uppercase_columns(df, column):
    df[column] = df[column].str.upper()

uppercase_columns(ga_full_applicants, 'name')
uppercase_columns(ga_full_applicants, 'address')
uppercase_columns(ga_full_applicants, 'city')
uppercase_columns(ga_full_applicants, 'business_type')
uppercase_columns(ga_full_applicants, 'loan_status')
uppercase_columns(ga_full_applicants, 'lender')
uppercase_columns(ga_full_applicants, 'servicing_lender_name')
uppercase_columns(ga_full_applicants, 'servicing_lender_address')
uppercase_columns(ga_full_applicants, 'servicing_lender_city')
uppercase_columns(ga_full_applicants, 'business_age_description')
uppercase_columns(ga_full_applicants, 'project_city')
uppercase_columns(ga_full_applicants, 'project_county_name')
uppercase_columns(ga_full_applicants, 'originating_lender_city')
```
